## Problem 1: Second largest element


```python
def second_largest(lst):
    if len(lst) < 2:
        return None
    largest = float('-inf')
    second = float('-inf')
    for num in lst:
        if num > largest:
            second = largest
            largest = num
        elif num > second and num != largest:
            second = num
    return second if second != float('-inf') else None

nums = [10, 20, 4, 45, 99, 99]
print(f'List: {nums}')
print(f'Second largest: {second_largest(nums)}')
```

## Problem 2: Frequency of each character


```python
def count_frequency(s):
    freq = {}
    for char in s:
        freq[char] = freq.get(char, 0) + 1
    return freq

text = 'hello world'
print(f"String: '{text}'")
print(f"Frequency: {count_frequency(text)}")
```

## Problem 3: Common elements without set


```python
def find_common(list1, list2):
    common = []
    for item in list1:
        if item in list2 and item not in common:
            common.append(item)
    return common

l1 = [1, 2, 3, 4, 5]
l2 = [4, 5, 6, 7, 8]
print(f"List 1: {l1}")
print(f"List 2: {l2}")
print(f"Common elements: {find_common(l1, l2)}")
```

## Problem 4: Armstrong Number


```python
def is_armstrong(n):
    num_str = str(n)
    power = len(num_str)
    total = sum(int(digit) ** power for digit in num_str)
    return total == n

num = 153
print(f"Is {num} an Armstrong number? {is_armstrong(num)}")
```

## Problem 5: Sort tuples by marks


```python
def sort_by_marks(students):
    return sorted(students, key=lambda x: x[1])

data = [('Alice', 85), ('Bob', 92), ('Charlie', 78)]
print(f"Original: {data}")
print(f"Sorted: {sort_by_marks(data)}")
```

## Problem 6: First non-repeating character


```python
def first_non_repeating(s):
    freq = {}
    for char in s:
        freq[char] = freq.get(char, 0) + 1
    for char in s:
        if freq[char] == 1:
            return char
    return None

text = 'swiss'
print(f"String: '{text}'")
print(f"First non-repeating character: {first_non_repeating(text)}")
```

## Problem 7: Remove duplicates maintaining order


```python
def remove_duplicates(lst):
    result = []
    seen = set()
    for item in lst:
        if item not in seen:
            seen.add(item)
            result.append(item)
    return result

nums = [1, 2, 2, 3, 4, 4, 5]
print(f"Original: {nums}")
print(f"Without duplicates: {remove_duplicates(nums)}")
```

## Problem 8: Swap keys and values


```python
def swap_dict(d):
    return {v: k for k, v in d.items()}

my_dict = {'a': 1, 'b': 2, 'c': 3}
print(f"Original: {my_dict}")
print(f"Swapped: {swap_dict(my_dict)}")
```

## Problem 9: Find target sum pairs


```python
def find_pairs(lst, target):
    pairs = []
    seen = set()
    for num in lst:
        complement = target - num
        if complement in seen:
            pairs.append((complement, num))
        seen.add(num)
    return pairs

nums = [1, 5, 7, -1, 5]
target_sum = 6
print(f"List: {nums}, Target: {target_sum}")
print(f"Pairs: {find_pairs(nums, target_sum)}")
```

## Problem 10: Flatten nested list


```python
def flatten_list(nested):
    flat = []
    for item in nested:
        if isinstance(item, list):
            flat.extend(flatten_list(item))
        else:
            flat.append(item)
    return flat

nested_data = [1, [2, [3, 4], 5], 6]
print(f"Nested: {nested_data}")
print(f"Flattened: {flatten_list(nested_data)}")
```
